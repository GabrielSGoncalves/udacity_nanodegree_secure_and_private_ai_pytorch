```python
import pandas as pd
```

```python
# Create Dataframe from CSV with dog breeds and weights
df = pd.read_csv('dog_breeds_weight_filtered.tsv', sep='\t')
```

```python
def get_chef_bob_plan(dog_weight_kg):
    chef_bob_plans = {
    'até 5kg': (0,5),
    'de 5 a 10kg': (5,10),
    'de 10 a 14kg': (10,14),
    'de 14 a 20kg': (14,20),
    'de 20 a 28kg': (20,28),
    'de 28 a 40kg': (28,40)}
    
    plan = 'Cachorro gigante!'
    for k,v in chef_bob_plans.items():
        if int(dog_weight_kg) in range(v[0],v[1]):
            plan = k
    return plan
```

```python
df['avg_male_weight_kg'] = round(df['avg_male_weight']/2.2, 2)
```

```python
df['avg_female_weight_kg'] = round(df['avg_female_weight']/2.2, 2)
```

```python
df['avg_weight_kg'] = round((df.avg_male_weight_kg + df.avg_female_weight_kg) / 2, 2)
```

```python
df['chef_bob_plan'] = df['avg_weight_kg'].apply(get_chef_bob_plan)
```
